```python
import matplotlib.pyplot as plt
```

```python
company=['samsung','Iplone','vivo','oppo','realme']
shares=[2002,2500,3500,2100,1800.3300]
```

```python
plt.figure(figsize=(18,4))
plt.plot(company,shares,color='g',linestyle="--",label='Mobile phone shares',marker='o')
plt.legend()
plt.xlabel(' Mobile Phone companies ')
plt.ylabel('shares')
plt.show()
```

```python
plt.figure(figsize=(18,4))
plt.plot(company,shares,color='g',linestyle="--",label='Mobile phone shares',marker='o')
plt.plot(company,[2300,3800,1700,3000,1500],color='k',linestyle="--",label='Mobile phone shares',marker='o')
plt.legend()
plt.xlabel(' Mobile Phone companies ')
plt.ylabel('shares')
plt.show()

```

```python
plt.figure(figsize=(18,4))
plt.bar(company,shares,color=['g','k','c','m','y'],label='Mobile phone shares')
plt.legend()
plt.xlabel(' Mobile Phone companies ')
plt.ylabel('shares')
plt.xticks(range(5),['SAMSUNG','IPHONE','VIVO','OPPO','REALME'],rotation=45)
plt.show()

```

```python
plt.figure(figsize=(11,4))
plt.barh(company,shares,color=['g','k','c','m','y'],label='Mobile phone shares')
plt.legend()
plt.ylabel(' Mobile Phone companies ')
plt.xlabel('shares')
plt.xticks(range(5),['SAMSUNG','IPHONE','VIVO','OPPO','REALME'],rotation=45)
plt.show()
```
