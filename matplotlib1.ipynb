```python
import matplotlib.pyplot as plt
plt

```

```python
days=['feb 5','feb 10','feb 15','feb 20','feb 25']
data=[100,55,50,25,15]
```

```python
plt.figure (figsize=(18,3))
plt.plot(days,data,linestyle='--',label='Tata Motors Shares',marker='o');
plt.legend()
plt.title('Share Price progress chart')
plt.ylabel('Share price')
plt.xlabel('Days of feb 2024')
```

```python
plt.figure (figsize=(18,3))
plt.bar(days,data,linestyle='--',label='Tata Motors Shares')
plt.legend()
plt.title('Share Price progress chart')
plt.ylabel('Share price')
plt.xlabel('Days of feb 2024')
```

```python
plt.figure (figsize=(18,3))
plt.bar(days,data,color=['c','m','y','k','c'],linestyle='--',label='Tata Motors Shares')
plt.legend()
plt.title('Share Price progress chart')
plt.ylabel('Share price')
plt.xlabel('Days of feb 2024')
```

```python
plt.figure (figsize=(18,3))
plt.barh(days,data,color=['c','m','y','k','c'],linestyle='--',label='Tata Motors Shares')
plt.legend()
plt.title('Share Price progress chart')
plt.xlabel('Share price')
plt.ylabel('Days of feb 2024')
```
