```python
import matplotlib.pyplot as plt
plt
```

```python
tod=['morning','afternoon','evening','night']
press=[150,110,170,100]
```

```python
plt.figure(figsize=(18,5))
plt.plot(tod,press,color='k',linestyle='--',label='compressed air ',marker='+');
plt.legend()
plt.xlabel('PARTS OF DAYS')
plt.ylabel('PRESSURISED AIR IN BAR')
```

```python
plt.figure(figsize=(18,5))
plt.bar(tod,press,color=['k','g','r','c'],linestyle='--',label='compressed air ');
plt.legend()
plt.xlabel('PARTS OF DAYS')
plt.ylabel('PRESSURISED AIR IN BAR')
```
